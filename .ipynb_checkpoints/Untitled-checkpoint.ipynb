```python
player1 = 'x'
player2 = 'o'

#is there any room on the board for anyone to move
def isMovePossible(board):
    for i in range(len(board)):
        for j in range(len(board)):
            if(board[i][j] == ' '):
                return True
    return False

def checkWinCondition(board):
    #check diagonals
    if (board[0][0] == board[1][1] and board[1][1] == board[2][2]):
        #player 1
        if(board[0][0] == 'x'):
            return 10
        #player 2 
        if(board[0][0] == 'o'):
            return -10
    if (board[0][2] == board[1][1] and board[1][1]== board[2][0]):
        #player 1
        if(board[1][1] == 'x'):
            return 10
        #player 2 
        if(board[1][1] == 'o'):
            return -10
    #check row/columns
    for i in range(0,3):
        #check each row for matching 3 set
        if (board[i][0] == board[i][1] and board[i][1] == board[i][2]):
            #player 1
            if(board[i][0] == 'x'):
                return 10
            #player 2 
            if(board[i][0] == 'o'):
                return -10
        #check each column
        if (board[0][i] == board[1][i] and board[1][i] == board[2][i]):
            #player 1
            if(board[0][i] == 'x'):
                return 10
            #player 2 
            if(board[0][i] == 'o'):
                return -10
    
        #no winners yet
        return 0
def minimax(board, depth, player):
    score = checkWinCondition(board)
    print(board,player)
    #player 1 win
    if(score == 10):
        return score
    #player 2 win
    if(score == -10):
        return score
    #is Tie no move possible and no one won
    if(isMovePossible(board) == False):
        return 0
    #checking for player1's best move score
    if(player):
        #is negative since we want a higher score
        best = -1000
        
        for i in range(len(board)):
            for j in range(len(board)):
                if(board[i][j] == ' '):
                    board[i][j] = 'x'
                    print('test:',board,player,i,j)
                    #recursively calls  and gets the maximum score
                    best = max(best, minimax(board, depth+1,not player))
                    board[i][j] = ' '
        return best
    else:
        #positive to get lowest score
        best = 1000 
        for i in range(len(board)):
            for j in range(len(board)):
                if(board[i][j] == ' '):
                    board[i][j] = 'o'
                    #recursively calls  and gets the minimum score
                    best = min(best, minimax(board, depth+1,not player))
                    board[i][j] = ' '
        return best
def findBestMove(board,player):
    bestMove = [-1,-1]
    if(player == 1):
        bestVal = -1000
    if(player == 2):
        bestVal = 1000
    for i in range(len(board)):
            for j in range(len(board)):
                if(board[i][j] == ' '):
                    if(player == 1):
                        board[i][j] = 'o'
                        moveVal = minimax(board, 0, True)
                        board[i][j] = ' '
                        if (moveVal > bestVal):
                            bestMove = [i,j]
                            bestMove = moveVal
                    else:
                        board[i][j] = 'x'
                        moveVal = minimax(board, 0, False)
                        board[i][j] = ' '
                        if (moveVal < bestVal):
                            bestMove = [i,j]
                            bestMove = moveVal
    print('Best Value is:',bestVal)
    return bestMove

def main():
    board = [['x','o','x'],
             ['x','o','x'],
             [' ',' ',' ']]
    bestMove = findBestMove(board,1)
    print(bestMove)
main()
```
